```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# constantes

Su = 1700e6
Sy = 1530e6
Fa = 5e3/2
k = 35e3
L = 1.5/2
E = 200e9
F_max = 5e3
F_min = 5e3

# plotagem grfico de goodman

fig = plt.figure(figsize=[16, 9])
fig.suptitle('Diagrama de Goodman', fontsize=16)

linha_sobrecarga_1 = np.array(
    ([0,0.45*Su],
    [Su,0])
) 

linha_sobrecarga_2 = np.array(
    ([0,Sy],
    [Sy,0])
) 

linha_carga = np.array(
    ([0,0],
    [Su,Su/2])
)


# Plotando 2D

ax = fig.add_subplot(1, 1, 1)

ax.plot(linha_sobrecarga_1[:,0], linha_sobrecarga_1[:,1], 'r', linewidth=2)
ax.plot(linha_sobrecarga_2[:,0], linha_sobrecarga_2[:,1], 'r', linewidth=2)
ax.plot(linha_carga[:,0], linha_carga[:,1], 'b', linewidth=2)

ax.set_xlabel("$tau_m$")
ax.set_ylabel("$tau_a$")

ax.set_xlim(0)
ax.set_ylim(0)

ax.grid()

plt.show()
```

```python
tensao_a_max = 0.4e9
print("{} MPa".format(tensao_a_max*1e-6))
```

```python
delta = 2*Fa/k
print("{:.4f} mm".format(delta*1e3))
```

```python
bh2 = 6*Fa*L/tensao_a_max
print('{:.4e}'.format(bh2))
```

```python
bh3 = 6*Fa*L**3/(E*delta)
print('{:.4e}'.format(bh3))
```

```python
h = bh2**0.5
print('{:.4e}'.format(h))
```

```python
b = h**3/bh3
print('{:.4f} mm'.format(b*1e3))
```
