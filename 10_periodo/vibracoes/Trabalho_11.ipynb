```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
```

```python
m = 500
k = 5000e3
c = 10e3
As = 0.002
sigma = 150
f0 = 745

omega = (k/m)**0.5

csi = c/(2*(k*m)**0.5)

B = f0/(m*omega)

raiz = (1-csi**2)**0.5

print(csi)
print(B)
print(raiz)

tau = np.linspace(0.000001,20,num=1000)

x = B * (np.exp(-csi*tau))/(raiz) * np.sin(raiz*tau)

print(np.max(x))

sigma0 = f0/As

print(sigma0*1e-6)
```

```python
fig = plt.figure(figsize=[16,9])
ax = fig.subplots(1,1)

ax.set_ylabel('$X( \tau )$',fontsize=16)
ax.set_xlabel('$ \tau $',fontsize=16)
ax.set_title('Gráfico do deslocamento $X( \tau)$ em função de $\tau$',fontsize=20)

ax.plot(tau,x ,'r', linewidth=4)

ax.grid()

plt.show()
fig.savefig('trabalho_11.png', bbox_inches='tight')
```
