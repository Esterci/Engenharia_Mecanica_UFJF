```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
```

```python
zeta = np.linspace(0.000001,0.7,num=1000)
phi = np.arctan(np.sqrt(1-zeta**2)/zeta)

a = 2*zeta*np.sqrt(1-zeta**2)
b = 1-2*zeta**2
psi = np.arctan(a/b)

F_b = np.exp((psi-phi)/np.tan(phi))
```

```python
fig = plt.figure(figsize=[16,9])
ax = fig.subplots(1,1)

ax.set_ylabel('$F_{b,max}/f_0 \omega _n$',fontsize=16)
ax.set_xlabel('$\zeta$',fontsize=16)
ax.set_title('Gráfico do angulo $F_{b,max}/f_0 \omega _n$ em função de $\zeta$',fontsize=20)

ax.plot(zeta,F_b ,'r', linewidth=4)

ax.grid()

plt.show()
fig.savefig('angulo_vs_tempo.png', bbox_inches='tight')
```

```python
phi = phi[np.argmin(F_b)]
zeta = zeta[np.argmin(F_b)]

print('Força {:.4f}'.format(np.min(F_b)))
print('zeta {:.4f}'.format(np.min(zeta)))
print('phi {:.4f}'.format(phi))
```

```python
v_0 = 3
m = 1e3
x_max = 0.5

omega_n = (v_0/x_max)*np.exp(-phi/np.tan(phi))

k = omega_n**2*m

c = zeta *2*np.sqrt(k*m)

print('omega_n  = {:.4f}'.format(omega_n))
print('k = {:.4e}'.format(k))
print('c = {:.4e}'.format(c))
```
