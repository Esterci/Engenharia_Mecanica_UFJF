```python
import numpy as np
import pandas as pd
import math as mt
import matplotlib.pyplot as plt
import sympy
from sklearn.preprocessing import MinMaxScaler
```

# Caso 1: Senoide

```python
omega_min = 670*2*np.pi/60
omega_max = 1350*2*np.pi/60
omega_opt = 955*2*np.pi/60
m = 9.8
F1 = 100
k_1 = 10e3

k_2 = omega_opt**2 * m

print('k2 = {:.2f} N/m'.format(k2))

Omega_min = omega_min * (m/k_1)**0.5

Omega_max = omega_max * (m/k_1)**0.5

Omega_eval = np.linspace(Omega_min,Omega_max,num=200)

omega_r = (k_2/k_1)**0.5

a1 = 1 + 2 * omega_r**2

a2 = omega_r**2

D0_min = Omega_min**4 - a1 * Omega_min**2 + a2

D0_max = Omega_max**4 - a1 * Omega_max**2 + a2

H11_max = (omega_r**2 - Omega_max**2)/D0_max

H21_max = omega_r**2/D0_max

H11_min = (omega_r**2 - Omega_min**2)/D0_min

H21_min = omega_r**2/D0_min

print('H11_max = {:.2f}'.format(H11_max))
print('H21_max = {:.2f}'.format(H21_max))
print('H11_min = {:.2f}'.format(H11_min))
print('H21_min = {:.2f}'.format(H21_min))
```

```python
D0_eval = Omega_eval**4 - a1 * Omega_eval**2 + a2

H11_eval = (omega_r**2 - Omega_eval**2)/D0_eval

H21_eval = omega_r**2/D0_eval

fig = plt.figure(figsize=[16,9])
ax = fig.subplots(1,1)

ax.set_ylabel('$H_{j1}(\Omega)$',fontsize=16)
ax.set_xlabel('$\Omega$',fontsize=16)
ax.set_title('Gráfico da amplitude adimensional em\nfunção de $\Omega$ com absorvedor',fontsize=20)

ax.plot(Omega_eval,H11_eval,'g', linewidth=2, label = '$H11')
ax.plot(Omega_eval,H21_eval,'b', linewidth=2, label = '$H21')
ax.plot(Omega_eval,(H11_eval-H21_eval),'r', linewidth=2, label = '$H21')

ax.grid()
ax.legend()
plt.show()
fig.savefig('trabalho15_com.png', bbox_inches='tight')
```

```python
H11_eval = 1/(1-Omega_eval**2)

fig = plt.figure(figsize=[16,9])
ax = fig.subplots(1,1)

ax.set_ylabel('$H_{11}(\Omega)$',fontsize=16)
ax.set_xlabel('$\Omega$',fontsize=16)
ax.set_title('Gráfico da amplitude adimensional em\nfunção de $\Omega$ sem absorvedor',fontsize=20)

ax.plot(Omega_eval,H11_eval,'r', linewidth=2, label = '$H11')

ax.grid()
ax.legend()
plt.show()
fig.savefig('trabalho15_sem.png', bbox_inches='tight')
```
