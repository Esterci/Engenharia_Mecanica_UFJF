```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
```

```python
tempo = np.linspace(0.001,10,num=100)
alpha0 = 0.2
omega_n = 1.96
zeta = 0.35

alpha_max = alpha0/1.02 * np.exp(-zeta*omega_n*1.54) * np.sin(1.02*1.54)
alpha = alpha0/1.02 * np.exp(-zeta*omega_n*tempo) * np.sin(1.02*tempo)

print('alpha_max = {:.4f} rad'.format(alpha_max))
```

```python
fig = plt.figure(figsize=[16,9])
ax = fig.subplots(1,1)

ax.set_ylabel('$ alpha (rad)$',fontsize=16)
ax.set_xlabel('$t(s)$',fontsize=16)
ax.set_title('Gráfico da amplitude adimensional em\nfunção do tempo adimensional',fontsize=20)

ax.plot(tempo,alpha ,'r', linewidth=4)

ax.grid()

plt.show()
fig.savefig('angulo_vs_tempo.png', bbox_inches='tight')
```
