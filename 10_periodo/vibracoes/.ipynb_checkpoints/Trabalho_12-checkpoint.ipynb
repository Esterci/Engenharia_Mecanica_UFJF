# Caso 1: Senoide

```python
v = 60 
m1 = 1
m2 = 2
y0 = 50
tau = np.linspace(0,25,num=200)
zeta_eval = np.linspace(0.06,0.7,num=200)

csi_list = [0.06,0.25,0.7]

Hmb = [2*zeta*np.exp(-zeta*tau)*np.sin((1-zeta**2)**0.5*tau)/(1-zeta**2)**0.5 + 
       1 - np.exp(-zeta*tau)*np.sin((1-zeta**2)**0.5*tau + np.arctan((1-zeta**2)**0.5/zeta))/(1-zeta**2)**0.5 
       for zeta in csi_list]

fig = plt.figure(figsize=[16,9])
ax = fig.subplots(1,1)

ax.set_ylabel('$x(tau )/a$',fontsize=16)
ax.set_xlabel('$tau $',fontsize=16)
ax.set_title('Gráfico da amplitude adimensional em\nfunção do tempo adimensional',fontsize=20)

ax.plot(tau,Hmb[0],'r', linewidth=2, label = '$\zeta = 0.6$')
ax.plot(tau,Hmb[1],'b', linewidth=2, label = '$\zeta = 0.25$')
ax.plot(tau,Hmb[2],'g', linewidth=2, label = '$\zeta = 0.70$')

ax.grid()
ax.legend()
plt.show()
fig.savefig('trabalho12_2d.png', bbox_inches='tight')
```

```python
X, Y = np.meshgrid(tau, zeta_eval)

Z = (2*Y*np.exp(-Y*X)*np.sin((1-Y**2)**0.5*X)/(1-Y**2)**0.5 + 
    1 - np.exp(-Y*X)*np.sin((1-Y**2)**0.5*X + np.arctan((1-Y**2)**0.5/Y))/(1-Y**2)**0.5 )

fig = plt.figure(figsize=[16,9])

ax = plt.axes(projection='3d')
    
ax.set_title('Gráfico da amplitude adimensional em\nfunção do tempo adimensional e do fator de amortecimento',fontsize=20)
ax.set_ylabel('$\zeta$',fontsize=16)
ax.set_xlabel('$tau$',fontsize=16)
ax.set_zlabel('$x(tau)/a$',fontsize=16)
    
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                    cmap='viridis', edgecolor='none')

ax.grid()

plt.show()
fig.savefig('trabalho12_3d.png', bbox_inches='tight')
```
